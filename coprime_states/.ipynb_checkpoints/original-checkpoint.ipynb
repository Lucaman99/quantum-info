## Generating Prime States

Imports all of the necessary dependencies for performing the calulations and numerical simulations:

```python
import numpy as np
import scipy.linalg
import math
import random
from matplotlib import pyplot as plt
from toqito.states.entanglement.negativity import negativity
from toqito.states.entanglement.concurrence import concurrence
from tqdm import tqdm
```

The following block of code generates the state vector representing the prime state, for a given dimension (which is accepted as an argument to the function). Essentially, what this function does is assigns each number a basis vector. We then cycle through each pair of numbers in the set of integer pairs up to an including the dimension, testing whether they are corpime or not. If they are, then the two corresponding basis vectors are tensored together and added the to "final vector". Once each pair has been tested, the final vector is normalized and returned. In addition, the function takes `ones` as an argument, which is simply a Boolean value that speciifies if we want to include co-prime pairs that include the number $1$:

```python
# Generating prime states numerically, for a given dimension

def generate_prime_state(dimension, ones):
    
    final_vector = np.array([0 for i in range(0, dimension**2)])
    
    for i in range(1, dimension+1):
        for j in range(1, dimension+1):
            
            if (not ones):
                if (math.gcd(i, j) == 1 and j != 1 and i != 1):

                    vector1 = np.array([0 for k in range(0, dimension)])
                    vector2 = np.array([0 for k in range(0, dimension)])

                    vector1[i-1] = 1
                    vector2[j-1] = 1

                    final_vector += np.kron(vector1, vector2)
            
            else:
                if (math.gcd(i, j) == 1):

                    vector1 = np.array([0 for k in range(0, dimension)])
                    vector2 = np.array([0 for k in range(0, dimension)])

                    vector1[i-1] = 1
                    vector2[j-1] = 1

                    final_vector += np.kron(vector1, vector2)
            
    final_vector = final_vector / np.sqrt(sum(final_vector))
    
    return final_vector
```

We can test our prime-state generating function for the case of dimension $D \ = \ 3$, with no ones:

```python
result = generate_prime_state(3, False)
print(result)
```

This is the result we expect, as we see there are non-zero entries in our vector corresponding to the pairs $(2, \ 3)$ and $(3, \ 2)$.

Next, we wish to construct a function that gives us the density matrix of a given quantum mechanical state vector. Since we are dealing with pure states, we simply have $\rho \ = \ | \psi \rangle \langle \psi |$, which can be computed easily:

```python
# Calculates the partial trace of the matrix

def density_matrix(vector):

    return np.outer(vector, np.conj(vector))
```

Now, we define a function that gives us the partial trace of a given density matrix. We do this by taking the trace of all the "subsections" that would correspond to density matrix of one half of our bipartite quantum system. Basically, if our quantum system can be represented by $\rho_{AB}$, then if $\rho_{AB}$ is separable, where $\rho_{AB} \ = \ \rho_{A} \ \otimes \ \rho_{B}$, then the trace of each subsection will correspond to the trace of $\rho_B$. Thus, since a density matrix has a partial trace of $1$, then taking the partial trace of a separable density matrix should simply yield $\rho_A$. Obviously, it is not always the case that our quantum system is separable, but this provides some motivation for the definition of a partial trace. This is not too difficult to implement as a function that takes a density matrix as an argument, and returns the partially-traced density matrix:

```python
def partial_trace(matrix):
    
    number = int(np.sqrt(len(matrix)))
    new_matrix = []
    
    for i in range(0, number):
        new_row = []
        for j in range(0, number):
            
            sub = matrix[i*number:(i+1)*number, j*number:(j+1)*number]
            new_row.append(np.trace(sub))
        
        new_matrix.append(new_row)
    
    return new_matrix
```

Now, let's test this on a simple density matrix:

```python
m = np.array([[1, 1, 2, 1], [1, 1, 1, 1], [4, 1, 1, 1], [1, 1, 1, 1]])
print(partial_trace(m))
```

This is the expected result!

Finally, in order to find the entropy of entanglement, we need to create a function that returns the Von Neumann entropy of a density matrix (specifically, the Von Neumann entropy of the reduced density matrix of the co-prime state in question). For a given density matrix $\rho$, the Von Neumann entropy is defined as $S(\rho) \ = \ - \rho \log \rho$, which can be implemented as follows:

```python
# Finds the Von Neumann entropy

def entropy(matrix):
    
    log = scipy.linalg.logm(matrix)
    final_val = -1*np.trace(np.matmul(matrix, log))
    return final_val
```

Finally, we compute the normalized Entropy of Entanglement (the Von Neumann entropy of the reduced density matrix) for all dimensions from $3$ to $50$:

```python
dimensions = range(3, 50)
y = []
y2 = []
for i in dimensions:
    
    prime_state = generate_prime_state(i, False)
    dm = density_matrix(prime_state)
    new_dm = partial_trace(dm)
    y.append(entropy(new_dm) / np.log(i))
    
    prime_state = generate_prime_state(i, True)
    dm = density_matrix(prime_state)
    new_dm = partial_trace(dm)
    y2.append(entropy(new_dm) / np.log(i))

# Creates the plot for co-prime states not containing ones
plt.plot(dimensions, y)
plt.show()

# Creates the plot for co-prime states containing ones
plt.plot(dimensions, y2)
plt.show()
```

Now that we have the graphs for the Entropy of Entanglement, we can check another entanglement metric, which is the value of the maximum eigenvalue of the reduced density matrix. Closer to $1$ indicates that the state is closer to separable. First, we can create a function that returns the largest eigenvalue of our reduced density matrix:

```python
# Finds the largest eigenvalue of the reduced density matrix

def largest_eigenval(matrix):
    
    return max(np.linalg.eig(matrix)[0])
```

Next, we can plot the values of the eigenvalues for density matrices ranging from dimension $3$ to $50$:

```python
dimensions = range(3, 50)
y = []
y2 = []

for i in dimensions:
    
    prime_state = generate_prime_state(i, False)
    dm = density_matrix(prime_state)
    new_dm = partial_trace(dm)
    y.append(largest_eigenval(new_dm))
    
    prime_state = generate_prime_state(i, True)
    dm = density_matrix(prime_state)
    new_dm = partial_trace(dm)
    y2.append(largest_eigenval(new_dm))

# Creates the plot for co-prime states not containing ones
plt.plot(dimensions, y)
plt.show()

# Creates the plot for co-prime states containing ones
plt.plot(dimensions, y2)
plt.show()
```
